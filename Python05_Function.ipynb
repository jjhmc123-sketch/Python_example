```python
def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5 / 9

temps = [77, 95, 50]

for t in temps:
    print(fahrenheit_to_celsius(t))
```

```python
def my_function(person):
    if "name" in person and "age" in person:
        print("Name:", person["name"])
        print("Age:", person["age"])
    else:
        print("Invalid data")

my_person = {"name": "Emil", "age": 25}
my_function(my_person)
```

```python
def my_function(*numbers):
    if len(numbers) == 0:
        return 0

    total = 0
    for num in numbers:
        total += num
    return total

print(my_function(1, 2, 3))
print(my_function())
```

```python
def my_function(*numbers):
    if len(numbers) == 0:
        return None

    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num

print(my_function(3, 7, 2, 9, 1))
```

```python
def my_function(**kid):
    if "lname" in kid:
        print("His last name is " + kid["lname"])
    else:
        print("lname not provided")

my_function(fname="Tobias", lname="Refsnes")
```

```python
def my_function(title, *args, **kwargs):
    print("Title:", title)

    for arg in args:
        print("Arg:", arg)

    for key, value in kwargs.items():
        print(key + ":", value)

my_function("User Info", "Emil", "Tobias", age=25, city="Oslo")
```

```python
def countdown(n):
    if n <= 0:
        print("Done!")
        return
    print(n)
    countdown(n - 1)

countdown(5)
```

```python
def fibonacci(n):
    if n < 0:
        return None
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(7))
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()

for value in gen:
    print(value)
```
